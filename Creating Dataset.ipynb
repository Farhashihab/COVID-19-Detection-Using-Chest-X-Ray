```python
import pandas as pd
import numpy as np
import os
datapath_1 = r'F:\\Covid-19 Detection\\chestxray'
dataset_path = r'F:\\Covid-19 Detection\\DataSet2'
categories = os.listdir(dataset_path)
# print(categories)
dataset = pd.read_csv(os.path.join(datapath_1,'metadata.csv'))
findings = dataset['finding']
views = dataset['view']
print(findings[-8:])
print(views)
image_names = dataset['filename']
# print(image_names)
```

# let's extract all the image index belongs to covid19 and sars

```python
positiveIndex = np.concatenate((np.where(findings[-8:]=='COVID-19')[0],np.where(views=='PA')[0]))
# print(positiveIndex)
positiveImageNames = image_names[positiveIndex]
positiveImageNames
```

## Extracting the images now

```python
import cv2

for names in positiveImageNames:
    image = cv2.imread(os.path.join(datapath_1,'images',names))
    try:
        cv2.imwrite(os.path.join(dataset_path,categories[1],names),image)
    except Exception as e:
        print(e)
```
